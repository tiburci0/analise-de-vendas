```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

Ler os dados

```python
df = pd.read_csv('../data/vendas.csv')
df['data'] = pd.to_datetime(df['data'])
df['valor_total'] = df['quantidade'] * df['preco_unitario']
df.head()
```

Total vendido por vendedor

```python
vendas_vendedor = df.groupby('vendedor')['valor_total'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sbn.barplot(x=vendas_vendedor.index, y=vendas_vendedor.values)
plt.title('Total vendido por vendedor')
plt.ylabel('Valor R$')
plt.xlabel('Vendedor')
plt.show()
```

Produtos mais vendidos

```python
produtos = df.groupby('produto')['quantidade'].sum().sort_values(ascending=True)

produtos.plot(kind='barh', figsize=(8,4), color='skyblue')
plt.title('Produtos mais vendidos')
plt.xlabel('Quantidade')
plt.show()
```
